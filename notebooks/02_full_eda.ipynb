```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
fraud_data = pd.read_csv('../data/Fraud_Data.csv')
ip_to_country = pd.read_csv('../data/IpAddress_to_Country.csv')
credit_card = pd.read_csv('../data/creditcard.csv')
fraud_data['signup_time'] = pd.to_datetime(fraud_data['signup_time'])
fraud_data['purchase_time'] = pd.to_datetime(fraud_data['purchase_time'])
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
fraud_data['class'].value_counts().plot(kind='bar', ax=ax[0], title='Fraud vs Non-Fraud (E-commerce)')
credit_card['Class'].value_counts().plot(kind='bar', ax=ax[1], title='Fraud vs Non-Fraud (Credit Card)')
plt.show()
sns.histplot(fraud_data['purchase_value'], bins=50, kde=True)
plt.title('Purchase Value Distribution (E-commerce)')
plt.show()

sns.histplot(fraud_data['age'], bins=30, kde=True)
plt.title('Age Distribution (E-commerce)')
plt.show()

sns.histplot(credit_card['Amount'], bins=50, kde=True)
plt.title('Transaction Amount Distribution (Credit Card)')
plt.show()
sns.boxplot(x='class', y='purchase_value', data=fraud_data)
plt.title('Purchase Value by Class (E-commerce)')
plt.show()

sns.boxplot(x='Class', y='Amount', data=credit_card)
plt.title('Amount by Class (Credit Card)')
plt.show()
fraud_data['hour_of_day'] = fraud_data['purchase_time'].dt.hour
fraud_data['day_of_week'] = fraud_data['purchase_time'].dt.dayofweek
fraud_data['time_since_signup'] = (fraud_data['purchase_time'] - fraud_data['signup_time']).dt.total_seconds() / 3600

sns.histplot(fraud_data['hour_of_day'], bins=24)
plt.title('Hour of Day Distribution (E-commerce)')
plt.show()

sns.histplot(fraud_data['day_of_week'], bins=7)
plt.title('Day of Week Distribution (E-commerce)')
plt.show()

sns.histplot(fraud_data['time_since_signup'], bins=50, kde=True)
plt.title('Time Since Signup (hours) Distribution (E-commerce)')
plt.show()
corr = credit_card.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap (Credit Card Data)')
plt.show()
print('E-commerce Data:')
display(fraud_data.describe())
print('Credit Card Data:')
display(credit_card.describe())
```
